```python
import numpy as np

# Caricamento del dataset
df = pd.read_csv('boston.csv')

# 1. Media del prezzo delle case in base alla distanza dal fiume Charles
media_prezzo_dist_charles = df.groupby('chas')['medv'].mean()
print("Media del prezzo delle case in base alla distanza dal fiume Charles:")
print(media_prezzo_dist_charles)

# 2. Correlazione tra il tasso di criminalità e il valore delle abitazioni
correlazione_crim_medv = df['crim'].corr(df['medv'])
print("Correlazione tra il tasso di criminalità e il valore delle abitazioni:", correlazione_crim_medv)

# 3. Media del numero di stanze rispetto al rapporto alunni-insegnanti e del valore delle case
media_stanze_ptratio = df['rm'].mean().round(2)
media_medv_ptratio = df['medv'].mean().round(2)
print("Media del numero di stanze:", media_stanze_ptratio)
print("Media del valore delle case:", media_medv_ptratio)

# 4. Variazioni della media delle età delle abitazioni rispetto all'accessibilità alle autostrade
media_eta_rad = df.groupby('rad')['age'].mean().round(0).astype(int)
print("Variazioni della media delle età delle abitazioni rispetto all'accessibilità alle autostrade:")
print(media_eta_rad)
```

```python

# Creazione del DataFrame dei dipendenti
employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id': [1, 2, 1, 2, 3]
})

# Creazione del DataFrame dei dipartimenti
departments_df = pd.DataFrame({
    'department_id': [1, 2, 3],
    'department_name': ['HR', 'IT', 'Finance'],
    'location': ['New York', 'San Francisco', 'Chicago']
})

# Esercizio 1/3: Unione dei DataFrame in base alla colonna comune 'department_id'
merged_df = pd.merge(employees_df, departments_df, on='department_id')

# Esercizio 2/3: Trasformazione della colonna 'department_id' in indice per entrambi i DataFrame
employees_df.set_index('department_id', inplace=True)
departments_df.set_index('department_id', inplace=True)

# Unione dei DataFrame mediante il metodo .join()
joined_df = employees_df.join(departments_df, how='inner')

# Esercizio 3/3: Verifica delle differenze nel risultato
print("Differenze tra i due DataFrame ottenuti:")
print(merged_df.equals(joined_df))
```
