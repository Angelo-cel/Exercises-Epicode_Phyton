```python
# Converti la colonna status_published in formato Timestamp
df['status_published'] = pd.to_datetime(df['status_published'])
df
```

```python
# Formatta le date delle pubblicazioni in un formato specifico
df['status_published_formatted'] = df['status_published'].dt.strftime("%Y-%m-%d %H:%M:%S")
df
```

```python
# Estraiamo solo i post relativi al 2012
posts_2012 = df[df['status_published'].dt.year == 2012]
posts_2012
```

```python
posts_may_2018 = df[(df['status_published'].dt.year == 2018) & (df['status_published'].dt.month == 5)]
posts_may_2018 
```

```python
# Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana
weekend_posts = df[df['status_published'].dt.dayofweek.isin([5, 6])]
weekday_posts = df[~df['status_published'].dt.dayofweek.isin([5, 6])]
num_weekend_posts = len(weekend_posts)
num_weekday_posts = len(weekday_posts)
print("Numero di post pubblicati nel weekend:", num_weekend_posts)
print("Numero di post pubblicati nel resto della settimana:", num_weekday_posts)
```

```python
# Troviamo il primo e ultimo post pubblicati in ogni anno
first_post_per_year = df.groupby(df['status_published'].dt.year)['status_published'].min()
last_post_per_year = df.groupby(df['status_published'].dt.year)['status_published'].max()
print(first_post_per_year)

print(last_post_per_year)
```

```python
import pandas as pd
df=pd.read_csv("automobile.csv")
df
```

```python
# Controlla se ci sono valori nulli nel dataset e conta quanti valori nulli ci sono in ogni colonna
null_values = df.isnull().sum()

# Stampa i valori nulli in ogni colonna
print("Valori nulli in ogni colonna:")
print(null_values)

```

```python
# Cerca le righe che hanno un valore nullo nella colonna num-of-doors
rows_with_null_doors = df[df['num-of-doors'].isnull()]

# Stampa le righe che hanno un valore nullo nella colonna num-of-doors
print("\nRighe con un valore nullo nella colonna num-of-doors:")
print(rows_with_null_doors)
```

```python
# Cerca le righe che hanno un valore nullo nella colonna num-of-doors
rows_with_null_doors = df[df['num-of-doors'].isnull()]
# Stampa le righe che hanno un valore nullo nella colonna num-of-doors
print("\nRighe con un valore nullo nella colonna num-of-doors:")
print(rows_with_null_doors)
```

```python
# Esamina i dati per trovare valori da sostituire ai valori nulli nella colonna num-of-doors
# Ad esempio, puoi sostituire i valori nulli con il valore più comune nella colonna num-of-doors
most_common_doors = df['num-of-doors'].mode()[0]
print("\nValore più comune nella colonna num-of-doors:", most_common_doors)

# Sostituisci i valori nulli nella colonna num-of-doors con il valore più comune
df['num-of-doors'].fillna(most_common_doors, inplace=True)

# Verifica se ci sono ancora valori nulli dopo la sostituzione
if df.isnull().sum().sum() == 0:
    print("\nNon ci sono più valori nulli nel dataset dopo la sostituzione.")
else:
    print("\nCi sono ancora valori nulli nel dataset.")
```

```python
import os
import pandas as pd

# Nome dell'utente
user_name = os.getlogin()

# Definisci il percorso relativo alla directory dei dataset
directory_path = r"C:\Users\{}\Desktop\EPICODE JUPYTER LAB\Classe jupyter\beginner datasets".format(user_name)

# Ottieni la lista dei file nella directory
file_list = os.listdir(directory_path)

# Inizializza una lista per memorizzare i nomi dei file con dati nulli
files_with_null_data = []

# Ciclo per esaminare ciascun file
for file_name in file_list:
    # Verifica se il file ha estensione .csv
    if file_name.endswith('.csv'):
        # Costruisci il percorso completo del file
        file_path = os.path.join(directory_path, file_name)
        
        # Carica il file in un DataFrame
        df = pd.read_csv(file_path)
        
        # Controlla se ci sono dati nulli nel DataFrame
        if df.isna().sum().sum() > 0:
            # Se ci sono dati nulli, aggiungi il nome del file alla lista
            files_with_null_data.append(file_name)

# Stampare i nomi dei file con dati nulli
print("I seguenti file contengono dati nulli:")
for file_name in files_with_null_data:
    print(file_name)
```
