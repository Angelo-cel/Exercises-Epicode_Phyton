```python
under_20 = df_diabete.loc[df_diabete['Age (years)'] < 20]
age_20_30 = df_diabete.loc[(df_diabete['Age (years)'] >= 20) & (df_diabete['Age (years)'] < 30)]
age_30_40 = df_diabete.loc[(df_diabete['Age (years)'] >= 30) & (df_diabete['Age (years)'] < 40)]
age_40_50 = df_diabete.loc[(df_diabete['Age (years)'] >= 40) & (df_diabete['Age (years)'] < 50)]
over_50 = df_diabete.loc[df_diabete['Age (years)'] > 50]

# Stampiamo il numero di righe per ogni fascia di età
print("Numero di persone con età inferiore a 20:", len(under_20))
print("Numero di persone con età compresa tra 20 e 30:", len(age_20_30))
print("Numero di persone con età compresa tra 30 e 40:", len(age_30_40))
print("Numero di persone con età compresa tra 40 e 50:", len(age_40_50))
print("Numero di persone con età superiore a 50:", len(over_50))
```

```python


import pandas as pd

# Effettua una copia del DataFrame
df_diabete_copy = df_diabete.copy()

# Definisci i bin e le etichette per le fasce di età
bins = [0, 20, 30, 40, 50, float('inf')]
labels = ['<20', '20-30', '30-40', '40-50', '>50']

# Crea una nuova colonna nel DataFrame che contiene le fasce di età
df_diabete_copy['Age Group'] = pd.cut(df_diabete_copy['Age (years)'], bins=bins, labels=labels, include_lowest=True, right=False)

# Calcola la media della pressione sanguigna diastolica per ciascuna fascia di età
media_pressione_per_eta = df_diabete_copy.groupby('Age Group', observed=False)['Diastolic blood pressure (mm Hg)'].mean()

# Stampiamo la media della pressione sanguigna diastolica per le diverse fasce di età
print("Media della pressione sanguigna diastolica per le diverse fasce di età:")
print(media_pressione_per_eta)
```

```python
import pandas as pd

# Calcoliamo la media della pressione sanguigna diastolica per le diverse fasce di età
media_pressione_per_fascia_eta = df_diabete.groupby(pd.cut(df_diabete['Age (years)'], bins=[0, 20, 30, 40, 50, float('inf')], include_lowest=True, right=False), observed=False)['Diastolic blood pressure (mm Hg)'].mean()

# Stampiamo la media della pressione sanguigna diastolica per le diverse fasce di età
print("Media della pressione sanguigna diastolica per le diverse fasce di età:")
print(media_pressione_per_fascia_eta)

# Calcoliamo la media della pressione sanguigna diastolica per ogni singolo anno di età
media_pressione_per_singolo_anno = df_diabete.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean()

# Stampiamo la media della pressione sanguigna diastolica per ogni singolo anno di età
print("\nMedia della pressione sanguigna diastolica per ogni singolo anno di età:")
print(media_pressione_per_singolo_anno)
```

```python
import pandas as pd
df_insurance = pd.read_csv("insurance.csv")
```

```python
df_insurance.columns
```

```python
df_insurance.dtypes
```

```python


# Calcola le medie delle spese mediche per ciascuna regione
medie_per_regione = df_insurance.groupby('region')['charges'].mean()

# Stampa le medie delle spese mediche per ciascuna regione
print("Medie delle spese mediche per ciascuna regione:")
print(medie_per_regione)
```

```python
medie_per_fumatore = df_insurance.groupby(['smoker',"region"])['charges'].mean()

# Stampa le medie delle spese mediche per ciascuno status di fumatore
print("Medie delle spese mediche per ciascuno status di fumatore:")
print(medie_per_fumatore)

```

```python
medie_per_sex = df_insurance.groupby(["sex","region"])["charges"].mean()
medie_per_sex
```

```python
descrittori_bmi = df_insurance['bmi'].describe()
print("Descrittori statistici del BMI:")
print(descrittori_bmi)

# Calcolo dei quartili del BMI
quartili_bmi = pd.qcut(df_insurance['bmi'], q=4)
df_insurance['bmi_quartile'] = quartili_bmi

# Calcolo dei valori minimi, medi e massimi delle spese mediche per ogni quartile di BMI
charges_per_quartile_bmi = df_insurance.groupby('bmi_quartile', observed=False)['charges'].agg(['min', 'mean', 'max'])
print("\nValori minimi, medi e massimi delle spese mediche per ogni quartile di BMI:")
print(charges_per_quartile_bmi)
```
